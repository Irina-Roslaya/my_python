```python
import random
listt=[1,2,3]
class Point: 
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
    def set(self, x, y):
        self.x = x
        self.y = y
    def __add__(self, other):
        return Point(self.x + other.x, self.y + other.y)
    def __mul__(self, other):
        return Point(self.x * other.x, self.y * other.y)
    def __str__(self):
        return f'({self.x},{self.y})'
class warrior():
    h=100
    def getdamage(self):
        self.h -= 20
    def showhealth(self):
        return self.h
class Superpowerman(warrior):
    h=100
    def getdamage(self):
        self.h -= 10
    def showhealth(self):
        return self.h 
warrior1 = warrior()
warrior2 = warrior()
warrior3 = Superpowerman()
pt1=Point(0,0)
pt1_1=Point(3, 0)
pt2=Point(0,0)
pt2_1 = Point(4, 0)
pt3=Point(0,0)
pt3_1 = Point(3, 5)
rock = Point(0.25,1)
rock2=Point(0.25,0.25)
leaf=Point(2,1)
leaf2=Point(2,2)
print("Начало\n При каждой стычке персонажи будут стараться убежать как можно дальше, на своем пути они могут встретить 2 предмета, которые распределяются между ними, в зависимости от свойств предмета меняется скорость передвижения персонажа до следующей стычки \n")
print("При обычных условиях 1 воин пробежит дистанцию: ",pt1_1,"При обычных условиях 2 воин пробежит дистанцию: ",pt2_1,"При обычных условиях воин с крыльями пролетит дистанцию: ",pt3_1)
print("Здоровье 1 воина: ", warrior1.showhealth(), "Здоровье 2 воина: ", warrior2.showhealth(), "Здоровье воина с крыльями: ", warrior3.showhealth())
print("Положение 1 воина: ",pt1,"Положение 2 воина: ",pt2,"Положение воина c крыльями: ",pt3)
round_of_fight = 0
while warrior1.h and warrior2.h and warrior3:
   
    if random.choice(listt)==1:
        warrior1.getdamage()
        pt1=pt1+pt1_1
        round_of_fight += 1
    elif random.choice(listt)==3:
        warrior3.getdamage()
        pt3=pt3+pt3_1
        round_of_fight += 1
    else:
        warrior2.getdamage()
        pt2=pt2+pt2_1
        round_of_fight += 1
    if  random.choice(listt)==1:
            print("Воин 1 взял замедляющий камень",",","Воин 2 взял волшебный ускоряющий листок",",","Воин с крыльями не взял ничего")
            pt1=pt1 + (pt1_1 * rock)
            pt2=pt2 + (pt2_1 * leaf)
    elif random.choice(listt)==2:
            print("Воин 2 взял замедляющий камень",",","Воин с крыльями взял волшебный ускоряющий листок",",","Воин 1 не взял ничего")
            pt2=pt2 + (pt2_1 * rock)
            pt3=pt3 + (pt3_1 * leaf2)
    else:
            print("Воин с крыльями взял замедляющий камень",",","Воин 1 взял волшебный ускоряющий листок",",","Воин 2 не взял ничего")
            pt3=pt3 + (pt3_1 * rock2)  
            pt1=pt1 + (pt1_1 * leaf)
    print("Количество стычек: {0}".format(round_of_fight))
    print("  Здоровье 1 воина: ", warrior1.showhealth(), "  Здоровье 2 воина: ", warrior2.showhealth(),"  Здоровье воина c крыльями: ", warrior3.showhealth())
    print("Положение 1 воина: ",pt1,"Положение 2 воина: ",pt2,"Положение воина c крыльями: ",pt3)
print()
if warrior1.h > warrior2.h and warrior1.h > warrior3.h :
    print("Первый воин победил")
elif warrior2.h > warrior1.h and warrior2.h > warrior3.h  :
    print("Второй воин победил")
else:
     print("Воин с крыльями победил")
print("Конец")
```
